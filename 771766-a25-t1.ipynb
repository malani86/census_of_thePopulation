**Step 2: Clean the Data**

* Age

```python
print(df.columns.tolist())
```

```python
#Strip whitespace from all column names
df.columns = df.columns.str.strip()

#Normalize all column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

print(df.columns.tolist())
```

```python
# Convert ageto numeric 
df['age'] = pd.to_numeric(df['age'], errors='coerce')


# Impute missing age by group
df['age'] = df.groupby(['relationship_to_head_of_house', 'gender'])['age'].transform(lambda x: x.fillna(x.median()))
```

* Marital Status

```python
# Fill any remaining missing ages with a reasonable default
df['age'] = df['age'].fillna(df['age'].median())

# Set marital_status to 'N/A' for anyone under 18
df.loc[df['age'] < 18, 'marital_status'] = 'N/A'


```

* Occupation Cleaning

```python
# Children
df.loc[df['age'] < 16, 'occupation'] = 'Student'

# Elderly
df.loc[(df['age'] > 70) & (df['occupation'].isnull()), 'occupation'] = 'Retired'

# Fill any remaining missing values
df['occupation'] = df['occupation'].fillna('Unemployed')
```

* Infirmity

```python
df['infirmity'] = df['infirmity'].fillna('None')
df['infirmity'] = df['infirmity'].str.strip().str.title()
```

* Religion

```python
df['religion'] = df['religion'].fillna('Not stated')
df['religion'] = df['religion'].str.strip().str.title()
```

* Gender

```python
df['gender'] = df['gender'].str.strip().str.title()
```

```python
# Check for any remaining missing values
print(df.isnull().sum())

# Check unique values (catch typos, variants)
for col in ['marital_status', 'occupation', 'infirmity', 'religion', 'gender']:
    print(f"{col}: {df[col].unique()}")
```

```python
df['relationship_to_head_of_house'] = df['relationship_to_head_of_house'].fillna('Unknown')

# Fix shorthand values
df['marital_status'] = df['marital_status'].replace({
    'D': 'Divorced',
    'W': 'Widowed',
    'S': 'Single',
    'M': 'Married'
})

# Fill remaining NaNs
df['marital_status'] = df['marital_status'].fillna('N/A')

df['infirmity'] = df['infirmity'].replace('', 'None')
df['infirmity'] = df['infirmity'].str.title().str.strip()

# First clean blanks
df['religion'] = df['religion'].replace('', 'Not Stated')

# Replace fake entries
invalid_religions = ['Housekeeper', 'Nope', 'Sith']
df['religion'] = df['religion'].replace(invalid_religions, 'Other')

# Standardize case
df['religion'] = df['religion'].str.title().str.strip()

df['gender'] = df['gender'].replace({'M': 'Male', 'F': 'Female'})
df['gender'] = df['gender'].str.title().str.strip()

```

```python
# Final null check
print(df.isnull().sum())

# Final value check
for col in ['marital_status', 'occupation', 'infirmity', 'religion', 'gender', 'relationship_to_head_of_house']:
    print(f"{col}: {df[col].unique()}")
```
